```python
from collections import Counter

grades = [83, 95, 91, 87, 70, 0, 85, 82, 100, 67, 73, 77, 0]

histogram = Counter(min(grade // 10 * 10, 90) for grade in grades)

plt.bar([x + 5 for x in histogram.keys()],
        histogram.values(),
        10,
        edgecolor=(1, 0, 0))

plt.axis([5, 105, 0, 5])

plt.xticks([10 * i for i in range(11)])
plt.xlabel("Decile")

plt.ylabel("# of Students")

plt.title("Distribution of Exam Grades")

plt.show()
```

```python
#Scatter-plot

nfriends = [70, 49, 10, 89, 300, 40, 98, 45, 90]
nminutes = [120, 200, 425, 50, 265, 100, 300, 78, 550]
labels = ["a", "b", 'c', "d", "e", "f", "g", "h", "k"]

plt.scatter(nfriends, nminutes)


for (label, friends, minutes) in zip(labels, nfriends, nminutes):
    plt.annotate(label, xy= (friends, minutes), xytext= (4, -4), textcoords="offset points")

# plt.axis("equal")
plt.title("Time Spent To # of Friends")
plt.xlabel("No of Friends")
plt.ylabel("Time Spent")

plt.show()
```

```python
#line Chart

variance = [1, 2, 4, 8, 16, 32, 64, 128, 256]
bias_squared = sorted(variance, reverse= True)
total_error = [x + y for x, y in zip(variance, bias_squared)]
xs = [i for i, _ in enumerate(variance)]

#plot xs against variance, bias_squared & total_error

plt.plot(xs, variance, "r-", label= "Variance")
plt.plot(xs, bias_squared, "g-",  label= "bias_squared")
plt.plot(xs, total_error, "b-", label= "total_error")

plt.legend(loc= 9)
plt.xticks([])
# plt.axis("equal")
plt.xlabel("Model Complexity")
plt.title("Bias-Variance TradeOff")
plt.show()
```

```python
from typing import List

Vector = List[float]

def add(v: Vector, w: Vector) -> Vector:
    assert len(v) == len(w), "VECTORS MUST BE OF SAME LENGTH"
    return [v_i + w_i for v_i, w_i in zip(v, w)]
assert add([1,2,3], [4,5,6]) == [5, 7, 9]


def subtract(v: Vector, w: Vector) -> Vector:
    assert len(v) == len(w), "Vectors must be of same length"
    return [v_i - w_i for v_i, w_i in zip(v, w)]

assert subtract([5, 7, 9], [4, 5, 6]) == [1, 2, 3]

def vector_sum(vectors: List[Vector]) -> Vector:
    assert vectors, "No vector provided"

    num_elements = len(vectors[0])
    assert all(len(v) == num_elements for v in vectors), "different sizes!"
    
    pass

```
