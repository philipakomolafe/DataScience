#from the above dataset we wanna comsider three things: HOST, MINIMUM NIGHT's & REVIEW-PER-MONTH
#Always Review www.matplotlib.org/gallery

```python
top_low_night_reviews

host_name, nights, monthly_review = top_low_night_reviews.host_name, top_low_night_reviews.minimum_nights, top_low_night_reviews.reviews_per_month

def do_plot(dataframe) -> any:
    host_names, nights, monthly_reviews = [x for x in dataframe.host_name], [x for x in dataframe.minimum_nights], [x for x in dataframe.reviews_per_month]
    
    # # plt.autoscale(enable=True, axis= "both")
    plt.figure(figsize= (15, 12))

    plt.bar(range(len(host_names)), monthly_reviews)

    plt.xticks(range(len(host_names)), host_names)


    plt.title("Comparison Between Reviews & Nights Spent")
    plt.xlabel("# of Nights")
    plt.ylabel("# of Monthly Reviews")
    plt.show()


do_plot(top_low_night_reviews)
top_low_night_reviews
```
